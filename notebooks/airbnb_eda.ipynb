## Data Cleaning

Small amounts of name and host names are empty. Since those are not significantly large, we can simply drop those rows.

We have about 20 percent of both "last_review" and "reviews_per_month" empty. It means these are new listings, with no review yet. We will need to either fill them, or drop the columns.

```python
# dropping rows

df = df.dropna(subset=["name", "host_name"])
```

```python
# Filling empty columns

df['reviews_per_month'] = df['reviews_per_month'].fillna(0)
df["last_review"] = df["last_review"].fillna("No Review.")
```

```python
# checking if everything worked

missing = df.isnull().sum()
print("missing values:\n", missing)
duplicates = df.duplicated().sum()
print("\n\nduplicates =", duplicates)
```

Now we have to change the "last review" column into the datetime, and have no problems with not a datetime rows

```python
df["last_review"] = pd.to_datetime(df["last_review"], errors = 'coerce')

print(df["last_review"].dtype)
print(df['last_review'].isna().sum())
```

## Now we have to deal with outliers.

```python
df = df[(df.price>=10) & (df.price<=1000)]
df.shape
```

```python
print(df['minimum_nights'].describe())
print(df[df['minimum_nights'] > 365].shape)
```

```python
df = df[df['minimum_nights'] <= 365]
df.shape
```

Now we have to drop columns, which are not helpful to us. Specifically id and host id

```python
df.drop("id", inplace=True, axis=1)
df.drop("host_id", inplace=True, axis=1)
df.columns
```

```python
print(df.shape)
print(df.isnull().sum())
```

## Adding new features

```python
#days since last review:

df["days_since_last_review"] = (pd.Timestamp('today') - df.last_review).dt.days
df['days_since_last_review'].describe()
```

```python
# host_listing_tier:

conditions = [
    df.calculated_host_listings_count == 1,
    df.calculated_host_listings_count < 6,
    df.calculated_host_listings_count >= 6 
]
choices = ['new', 'established', 'power']

df = df.reset_index(drop=True)

df['host_listing_tier'] = np.select(conditions, choices, default='unknown')

print(df['host_listing_tier'].value_counts())
```

```python
# review_activity:

df['review_activity'] = (df['reviews_per_month'] > 0).astype(int)

df['review_activity'].value_counts()
```

```python
#availability_rate:

df["availability_rate"] = (df['availability_365'] / 365).round(2)

df['availability_rate'].describe()
```

```python
df['price_per_room_type'] = df.groupby('room_type')['price'].transform('mean').round(2)
print(df.groupby('room_type')['price_per_room_type'].first())
```

Queens is a distant third

Bronx and Staten Island are nearly invisible at this scale

*Business insight*: Supply is heavily concentrated in two boroughs. A new host in Queens or the Bronx faces far less competition — worth flagging.

## Bivariate & Multivariate Analysis

We have four visualizations planned:

1. Price by neighbourhood group (boxplot)
2. Price by room type (boxplot)
3. Correlation heatmap
4. Host listing tier vs price vs reviews

```python
fig, ax = plt.subplots(figsize=(10,6))

sns.boxplot(data=df, x = 'neighbourhood_group', y = "price", 
            order= df.groupby('neighbourhood_group')["price"].median().sort_values(ascending=True).index)

ax.set_title('Neighbourhood vs Price')
ax.set_xlabel("neighbourhood_group")
ax.set_ylabel("price")

plt.tight_layout()
plt.savefig('../visuals/Neighbourhood vs Price.png', dpi=150, bbox_inches='tight')
plt.show()
```

Key insights from this chart:

- Manhattan has the highest median price and the widest IQR — most pricing diversity
- Bronx and Staten Island cluster tightly at the low end — affordable but limited supply
- Brooklyn is interesting — median is lower than Manhattan but has nearly as many high-price outliers, suggesting a split market of budget and premium listings
- Every borough has outliers reaching $1,000 — premium listings exist everywhere

```python
fig, ax = plt.subplots(figsize=(8,6))


sns.boxplot(data=df, x = 'room_type', y = "price", 
            order= df.groupby('room_type')["price"].median().sort_values(ascending=False).index)

ax.set_title('Room Type vs Price')
ax.set_xlabel("Room type")
ax.set_ylabel('Price')

plt.tight_layout()

plt.savefig('../visuals/Room Type vs Price.png', dpi=150, bbox_inches='tight')
plt.show()
```

Key insights:

- Entire home/apt median is roughly double Private room — confirms the price_per_room_type feature we built
- Shared room IQR is tight and low — very little pricing power in that category
- All three types have outliers reaching $1,000 — premium pricing exists regardless of room type

```python
corr_matrix = df.select_dtypes(include='number').corr()

fig, ax = plt.subplots(figsize=(12,10))
sns.heatmap(corr_matrix, 
            annot=True, 
            fmt=".2f",
            cmap="coolwarm",
            center = 0,
            ax = ax)

plt.tight_layout()
plt.savefig('../visuals/Heatmap', dpi=150, bbox_inches='tight')
plt.show()
```

### Host listing tier vs price vs reviews

```python
fig, axes = plt.subplots(1, 2, figsize=(14,6))
tier_order = ['new', 'established', 'power']

sns.boxplot(data=df, x = 'host_listing_tier', y = "price", 
            order= tier_order,
            ax = axes[0])
axes[0].set_title('Price by Host Tier')
axes[0].set_xlabel("Host Listing Tier")
axes[0].set_ylabel('Price')


sns.boxplot(data=df, x = 'host_listing_tier', y = "reviews_per_month",
             ax=axes[1],
             order= tier_order)
axes[1].set_title('Review Activity by Host Tier')
axes[1].set_xlabel('Host Listing Tier')

fig.suptitle('Host listing tier vs price vs reviews', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('../visuals/Host listing tier vs price vs reviews.png', dpi=150, bbox_inches='tight')
plt.show()
```

More listings doesn't automatically mean better performance. Established hosts (2–5 listings) appear to be the sweet spot — focused enough to maintain review activity, experienced enough to price competitively.

#### 1. What drives listing prices in NYC?

Room type is the strongest driver of price, with entire apartments commanding roughly double the rate of private rooms. Neighbourhood is the second key factor — Manhattan listings carry a consistent price premium over outer boroughs. Host tier also plays a role: surprisingly, new hosts price higher than established ones, possibly attempting to signal premium quality without the review history to justify it.

#### 2. Which neighbourhoods offer the best value?

Brooklyn offers the best overall value. Its median price is lower than Manhattan's, yet it produces a comparable number of high-price outliers — indicating a diverse market that accommodates both budget and premium guests.

#### 3. What separates high-performing hosts from low-performing ones?

Established hosts (2–5 listings) outperform both new and power hosts on review activity and recency. Power hosts, despite managing the most listings, receive fewer reviews per listing — suggesting bookings are spread too thin. The data indicates that focused, multi-property hosting is more effective than large-scale portfolio management.

#### 4. Where should a new host list their property to maximize bookings?

Queens, the Bronx, and Staten Island have significantly lower listing density than Manhattan and Brooklyn. A new host entering these markets faces less competition, making it easier to accumulate reviews and build a performance track record.
